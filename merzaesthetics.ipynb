```python
def generateSQL(clinic): 
    sql = "INSERT INTO `Merzthailand` " \
    "(`id`, `clinic_name_th`, `clinic_name_en`, `alias`, `xeomin_flag`, `belotero_flag`, `ulthera_flag`, `province`, `district`, `website`, `facebook`, `tel`, `latitude`, `longtitude`, `xeomin`, `x_beauty`, `belotero`, `skin_radiance`, `ulthera`) " \
    "VALUES ({}, '{}', '{}', '{}', '{}','{}', '{}', '{}', '{}', '{}', '{}', '{}' , {}, {}, '{}', '{}', '{}', '{}', '{}');"
    return sql.format(clinic['id'],
           clinic['name'],
           clinic['name_2'],
           urllib.parse.unquote(clinic['alias']),
           clinic['xeomin_flag'],
           clinic['belotero_flag'],
           clinic['ulthera_flag'],
           clinic['province'],
          clinic['district'],
          urllib.parse.unquote(clinic['website']),
          urllib.parse.unquote(clinic['facebook']),
          '/'.join(clinic['tel']),
          clinic['latitude'] or 'NULL',
          clinic['longitude'] or 'NULL',
          clinic['storeproducts'][0]['flag'],
          clinic['storeproducts'][1]['flag'],
          clinic['storeproducts'][2]['flag'],
          clinic['storeproducts'][3]['flag'],
          clinic['storeproducts'][4]['flag'])
    
clinic = results['data'][11]
sql = generateSQL(clinic)
sql
```

```python
for clinic in results['data']:
    try:
        sql = generateSQL(clinic)
        conn.query(sql)
        print(f"Insert! {clinic['name']}")
    except Exception as e:
        print('ERROR! {} {}'.format(clinic['name'], e))
    
```
