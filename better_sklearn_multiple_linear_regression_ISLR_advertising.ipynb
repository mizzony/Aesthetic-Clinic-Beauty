## sklearn: LinearRegression

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
print(f'sklearn version: {sklearn.__version__}')
```

```python
model = LinearRegression()
model
#create object from class LinearRegression
```

```python
len(df). #check how many row we have in df
```

```python
df.columns #checl column head
```

#try simply split 140 from 200 row as a train data set


```python
X=df.drop(columns=['Sales'])[:140] #drop Sales(Y) columns we will use as Y
y=df['Sales'][:140]
```

```python
len(X) #check X
```

```python
len(y)
```

```python
X.head()
```

```python
y.head() #return a series of id and value as below
```

```python
model.fit(X, y)
```

```python
model.score(X, y) # R-squared
```

```python
model.intercept_ #Y at x=0
```

```python
model.coef_ #will return array of B1, B2, B3 as how many X you have**
```

```python
X.head(3)
```

```python
model.predict([[200, 40, 70]]) #try use X1,x2,x3 like this for see the Y predicted
```

```python
model.predict([[200, 40, 70], 
               [100, 80, 50],
               [ 40, 20, 10]]) #predicted more than 1 value
```

```python
X_test=df.drop(columns=['Sales'])[140:]
X_test.head()
```

```python
y_hat=model.predict(X_test)
y_hat #use test data from 140> end (60 predicters/testers) as predictors then it will return numpy arrays like thihs
```

```python
dc=pd.concat([df[140:].reset_index(), pd.Series(y_hat, name='y_hat')], axis='columns')
dc #make beautiful Y_hat = predicted next to Y 
```

## Statsmodels: Multiple Linear regression

It will give a lot more detials than just scikitlearn lets see


```python
import statsmodels.api as sm
import statsmodels.formula.api as smf #for write equation very cool
```

```python
# formula: response ~ predictor1 + predictor2
model_a = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=df[:140]).fit() #write the quation like this and use
#use only first140 rows
# model_a = smf.ols(formula='Sales ~ TV + Radio', data=df).fit()
```

```python
print(model_a.summary())
```


1.  R, R adjusted should close to 1
