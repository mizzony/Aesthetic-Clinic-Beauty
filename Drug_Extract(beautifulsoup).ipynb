```python
from bs4 import BeautifulSoup 
import requests #for function to call via HTTPs from website see HTTP protocol
```

```python
res = requests.get("https://www.rxlist.com/drugs/alpha_a.htm")
soup = BeautifulSoup(res.text, "html.parser")  #initial soup by parseing HTML text element
#print(dir(res)) #checking directory of requests
print(res.status_code)  
```

```python
print(soup.prettify())  #to make code look pretty
```

```python
print(soup.title.text)  
```

```python
table = soup.find('div', {'class': 'AZ_results'}) #in this case find return arrays,find_all return vector
lis = table.find_all("li")
print(lis[0])
```

```python
all_drugs = []
for li in lis :
    all_drugs.append({'name' : li.find('a').text, 'fda' : li.find('span').text.replace('- ', '')})
    
print(all_drugs)
```

```python
import csv
all_drugs = []
character = [chr(i) for i in range(ord('a'), ord('z') + 1)]
print(character)
for aa in character : 
  res = requests.get("https://www.rxlist.com/drugs/alpha_{}.htm".format(aa))
  soup = BeautifulSoup(res.text, "html.parser")  #initial soup by parseing HTML text element

  table = soup.find('div', {'class': 'AZ_results'}) #in this case find return arrays,find_all return vector
  lis = table.find_all("li")
  for li in lis :
    all_drugs.append({'name' : li.find('a').text, 'fda' : li.find('span').text.replace('- ', '')})
    
print(all_drugs)

with open('all_drugs.csv', mode='w') as pc:
    file_writer = csv.writer(pc, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    file_writer.writerow(['name', 'fda'])
    for drugs in all_drugs :
        file_writer.writerow([drugs['name'], drugs['fda']])
```
