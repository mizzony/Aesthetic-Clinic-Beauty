```python
all_posts = []
for post in soup.find_all('li', {'class': 'pt-list-item'}) :
    res = requests.get(post.find('a').get('href'))
    soup2 = BeautifulSoup(res.content, "html.parser") #initial soup
    all_posts.append({
        'title': post.find('a').text,
        'post':  soup2.find('div', {'class': 'display-post-story'}).text.strip()
    })
```

```python
print(all_posts)
```
